# Data Preprocessing (nulls, duplicates, data format, ..etc)

```python

print(final_car_dataset['price'].head())
```

```python
# function to convert mixed units in common INR
def convert_price_to_inr(price):
    # remove the commas (,) and currency symbol (₹) and convert to lower case
    price = price.replace(',', '').replace('₹', '').strip().lower()
    
    # convert to base unit
    if 'crore' in price:
        return round(float(price.replace('crore', '').strip()) * 1e7, 2)
    elif 'lakh' in price:
        return round(float(price.replace('lakh', '').strip()) * 1e5, 2)
    else:
        # assume if already is in INR and not unit is present
        try:
            return round(float(price), 2)
        except ValueError:
            return None
    
```

```python
# price column
final_car_dataset['price'] = final_car_dataset['price'].apply(convert_price_to_inr)
final_car_dataset['price'].astype(float)
```

```python
final_car_dataset.head(2)
```

```python
# Body_Type 
final_car_dataset['Body_Type'] = final_car_dataset['Body_Type'].fillna(final_car_dataset['Body_Type'].mode()[0])
```

```python
# Insurance validity 
# Fill missing values in Insurance_Validity based on the mode 
final_car_dataset['Insurance_Validity'] = final_car_dataset['Insurance_Validity'].fillna(final_car_dataset['Insurance_Validity'].mode()[0])
```

```python
# Kms_Driven
final_car_dataset['Kms_Driven'] = final_car_dataset['Kms_Driven'].str.replace(",","").str.replace("Kms","").str.strip()
final_car_dataset['Kms_Driven'] = final_car_dataset['Kms_Driven'].astype(float)
```

```python
# fill the missing value of Kms_driven with average
final_car_dataset['Kms_Driven'] = final_car_dataset['Kms_Driven'].fillna(final_car_dataset['Kms_Driven'].mean())
```

```python
# Mileage 
# extract the 'kmpl' from mileage
final_car_dataset['Mileage'] = final_car_dataset['Mileage'].str.extract(r'([\d\.]+)').astype(float)
```

```python
# fill the null values of  mileage with mean()
final_car_dataset['Mileage'] = final_car_dataset['Mileage'].fillna(final_car_dataset['Mileage'].mean())
```

```python
# Seats column
# extract the numeric part in seats
final_car_dataset['Seats'] = final_car_dataset['Seats'].str.extract(r'([\d]+)').astype(float)
# fill the null value  by mean()
final_car_dataset['Seats'] = final_car_dataset['Seats'].fillna(final_car_dataset['Seats'].mean()).astype(int)
```

```python
# Engine column
# Extract the `CC` from the Engine
final_car_dataset['Engine'] = final_car_dataset['Engine'].str.extract(r'([\d]+)').astype(float)
# fill the null value by mean()
final_car_dataset['Engine'] = final_car_dataset['Engine'].fillna(final_car_dataset['Engine'].mean()).astype(int)
```

```python
# Color column
# fill the missing value by forward fill method
final_car_dataset['Color'] = final_car_dataset['Color'].ffill()
```

```python
# inspect the null values and duplicate and data format
final_car_dataset.isnull().sum()
```

```python
# drop the duplicates
final_car_dataset.drop_duplicates(inplace=True)
```
